```python
# Accès à l'ajout des données à la base :
# Créer une question

url = 'http://127.0.0.1:5000/api/v1/resources/questions/create_question'

nvl_question = {'id': 21,
     'question': 'Citer les differents types d’effets que peuvent avoir les mors',
     'type_question': 'Sans ordre',
     'niveau': 4}

create_question = requests.post(url, json = nvl_question)

print(create_question.text)

# La cellule renvoie un message d'erreur mais la commande fonctionne quand même : une nouvelle question est bien créée.
```

```python
# Accès à l'ajout des données à la base :
# Créer une réponse

url = 'http://127.0.0.1:5000/api/v1/resources/reponses/create_reponse'

nvl_reponse = {'id': 31,
     'reponse': "Effet releveur, effet abaisseur, effet d’encadrement, effet levier",
     'type_reponse': "Sans ordre",
     'niveau': 4,
     'question_id': 21} # Ajouter l'ID de la question à laquelle la réponse est liée

create_reponse = requests.post(url, json = nvl_reponse)

print(create_reponse.text)

# La cellule renvoie un message d'erreur mais la commande fonctionne quand même : une nouvelle réponse est bien créée.
```

```python
# Accès à la mise à jour des données dans la base :
# Mise à jour d'une question

url = 'http://127.0.0.1:5000/api/v1/resources/questions/update_question'

# Mise à jour de la question choisie :
upd_question = {'id': 21, # Exemple avec la question 21
     'question': 'Citer les types d effets que peuvent avoir les mors sur le cheval', # (Ajout de ”sur le cheval” à la question originale)
     'type_question': 'Sans ordre',
     'niveau': '4'}

maj_question = requests.put(url, json = upd_question)

print(maj_question.text)
```

```python
# Accès à la mise à jour des données dans la base :
# Mise à jour d'une réponse

url = 'http://127.0.0.1:5000/api/v1/resources/reponses/update_reponse'

# Mise à jour de la réponse choisie :
upd_reponse = {'id': 31, # Exemple avec la réponse 31
     'reponse': 'Releveur, abaisseur, encadrement, levier', # (Formulation de la réponse différente)
     'type_reponse': 'Sans ordre',
     'niveau': '4',
     'question_id': 21}

maj_reponse = requests.put(url, json = upd_reponse)

print(maj_reponse.text)
```

```python
# Suppression des données :
# Mise à jour d'une question (et de sa réponse associée/jointe)


url = 'http://127.0.0.1:5000/api/v1/resources/questions/delete_question'

id_demande = {'id': 21} # ID de la question qu'on souhaite supprimer

suppr_question = requests.delete(url, json= id_demande)

print(suppr_question.text)
```
